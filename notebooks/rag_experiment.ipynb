```python
import re

def preprocess_query(query):

    query = query.lower()

    query = re.sub(r"[^a-zA-Z0-9\s]", "", query)

    replacements = {
        "vts": "vessel traffic service",
        "aton": "aids to navigation"
    }

    for k, v in replacements.items():
        query = query.replace(k, v)

    return query
```

```python
from rank_bm25 import BM25Okapi

corpus = [doc.page_content for doc in chunks]

tokenized_corpus = [doc.split(" ") for doc in corpus]

bm25 = BM25Okapi(tokenized_corpus)

print("BM25 index created")
```

```python
query = preprocess_query("What is vessel traffic service")

tokenized_query = query.split(" ")

scores = bm25.get_scores(tokenized_query)

top_n = sorted(range(len(scores)), key=lambda i: scores[i], reverse=True)[:5]

for i in top_n:
    print(chunks[i].page_content[:300])
    print("\n")
```

```python
def hybrid_retrieval(query, k=10):

    query = preprocess_query(query)


    dense_docs = vectorstore.similarity_search(query, k)


    tokenized_query = query.split(" ")
    bm25_scores = bm25.get_scores(tokenized_query)

    bm25_top = sorted(
        range(len(bm25_scores)),
        key=lambda i: bm25_scores[i],
        reverse=True
    )[:k]

    bm25_docs = [chunks[i] for i in bm25_top]

    combined_docs = dense_docs + bm25_docs

    return combined_docs
```

```python
results = hybrid_retrieval("What is vessel traffic service?", k=5)

for i, doc in enumerate(results):
    print("Result", i+1)
    print(doc.page_content[:300])
    print("\n")
```

```python
from sentence_transformers import CrossEncoder

reranker = CrossEncoder(
    "cross-encoder/ms-marco-MiniLM-L-6-v2"
)

print("Cross encoder loaded")
```

```python
def rerank_documents(query, docs, top_n=5):

    pairs = [(query, doc.page_content) for doc in docs]

    scores = reranker.predict(pairs)

    ranked = sorted(
        zip(docs, scores),
        key=lambda x: x[1],
        reverse=True
    )

    top_docs = [doc for doc,_ in ranked[:top_n]]

    return top_docs
```

```python
def precision_at_3(pages):

    unique_pages = set(pages)

    return len(unique_pages) / 3
```

```python
def reciprocal_rank(pages):

    first_page = pages[0]

    for i, page in enumerate(pages):

        if page == first_page:
            return 1/(i+1)

    return 0
```
